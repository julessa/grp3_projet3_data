```python
import kagglehub
import os

# Téléchargement du dataset LIAR
path = kagglehub.dataset_download("doanquanvietnamca/liar-dataset")

print(f"✅ Dataset téléchargé dans : {path}")
print(f"\n📂 Contenu du dossier :")
for f in os.listdir(path):
    filepath = os.path.join(path, f)
    size = os.path.getsize(filepath) / 1024
    print(f"   - {f} ({size:.1f} KB)")
```

# Import des données avec Pandas

Le dataset est composé de **3 fichiers TSV** (Tab-Separated Values) :
- `train.tsv` — Données d'entraînement
- `test.tsv` — Données de test
- `valid.tsv` — Données de validation

Chaque fichier contient **14 colonnes** :

| # | Colonne | Description |
|---|---------|-------------|
| 1 | `id` | Identifiant unique de la déclaration |
| 2 | `label` | Étiquette de véracité (pants-fire, false, barely-true, half-true, mostly-true, true) |
| 3 | `statement` | La déclaration évaluée |
| 4 | `subject` | Sujet(s) de la déclaration |
| 5 | `speaker` | Auteur de la déclaration |
| 6 | `speaker_job_title` | Titre professionnel de l'auteur |
| 7 | `state_info` | État d'origine de l'auteur |
| 8 | `party_affiliation` | Affiliation politique |
| 9 | `barely_true_counts` | Nb de déclarations "barely-true" de l'auteur |
| 10 | `false_counts` | Nb de déclarations "false" de l'auteur |
| 11 | `half_true_counts` | Nb de déclarations "half-true" de l'auteur |
| 12 | `mostly_true_counts` | Nb de déclarations "mostly-true" de l'auteur |
| 13 | `pants_on_fire_counts` | Nb de déclarations "pants-fire" de l'auteur |
| 14 | `context` | Contexte / lieu de la déclaration |

```python
import pandas as pd
import os

# Noms des 14 colonnes du dataset LIAR
column_names = [
    'id',
    'label',
    'statement',
    'subject',
    'speaker',
    'speaker_job_title',
    'state_info',
    'party_affiliation',
    'barely_true_counts',
    'false_counts',
    'half_true_counts',
    'mostly_true_counts',
    'pants_on_fire_counts',
    'context'
]

# Chargement des 3 fichiers TSV
df_train = pd.read_csv(os.path.join(path, 'train.tsv'), sep='\t', header=None, names=column_names)
df_test  = pd.read_csv(os.path.join(path, 'test.tsv'),  sep='\t', header=None, names=column_names)
df_valid = pd.read_csv(os.path.join(path, 'valid.tsv'), sep='\t', header=None, names=column_names)

print(f"✅ Données chargées avec succès !")
print(f"   📊 Train : {df_train.shape[0]:,} lignes x {df_train.shape[1]} colonnes")
print(f"   📊 Test  : {df_test.shape[0]:,} lignes x {df_test.shape[1]} colonnes")
print(f"   📊 Valid : {df_valid.shape[0]:,} lignes x {df_valid.shape[1]} colonnes")
print(f"   📊 Total : {df_train.shape[0] + df_test.shape[0] + df_valid.shape[0]:,} lignes")
```

```python
# Sauvegarde au format CSV dans le dossier courant
df_train.to_csv('train.csv', index=False)
df_test.to_csv('test.csv', index=False)
df_valid.to_csv('valid.csv', index=False)

print("✅ Extraction terminée ! Les 3 fichiers (train.csv, test.csv, valid.csv) ont été créés.")
```
