```python
import pandas as pd
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../dataimport/train.csv")
valid = pd.read_csv("../dataimport/valid.csv")
test = pd.read_csv("../dataimport/test.csv")

print("Train :", train.shape)
print("Valid :", valid.shape)
print("Test  :", test.shape)
```

```python
print(train.columns)

print("\nValeurs manquantes dans train :")
print(train.isna().sum().sort_values(ascending=False))

print("\nValeurs manquantes dans valid :")
print(valid.isna().sum().sort_values(ascending=False))

print("\nValeurs manquantes dans test :")
print(test.isna().sum().sort_values(ascending=False))
```

```python
stop_words = set(ENGLISH_STOP_WORDS)

def clean_text(text):
    if pd.isna(text):
        return ""
    
    text = str(text).lower().strip()              # minuscule
    text = re.sub(r"http\S+|www\.\S+", " ", text) # supprime URLs
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)   # supprime ponctuation
    text = re.sub(r"\s+", " ", text).strip()      # espaces propres
    
    tokens = text.split()                         # tokenisation simple
    tokens = [tok for tok in tokens if tok not in stop_words]  # stopwords
    
    return " ".join(tokens)
```

```python
missing_train = (train.isna().sum() / len(train) * 100).sort_values(ascending=False)
missing_valid = (valid.isna().sum() / len(valid) * 100).sort_values(ascending=False)
missing_test = (test.isna().sum() / len(test) * 100).sort_values(ascending=False)

print("Pourcentage de NA - train")
print(missing_train)

print("\nPourcentage de NA - valid")
print(missing_valid)

print("\nPourcentage de NA - test")
print(missing_test)
```
